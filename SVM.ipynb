```python
import pandas as pd
from sklearn.svm import OneClassSVM
from sklearn.preprocessing import StandardScaler
from sklearn.impute import SimpleImputer

data = pd.read_csv('/content/financial_anomaly_data[1].csv') 

data_sample = data.sample(n=1000, random_state=42) 

numerical_columns = data_sample.select_dtypes(include=['int', 'float']).columns.tolist()
X = data_sample[numerical_columns]

imputer = SimpleImputer(strategy='mean')
X_imputed = imputer.fit_transform(X)

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X_imputed)

model = OneClassSVM(kernel='rbf', nu=0.05)  
model.fit(X_scaled)

predictions = model.predict(X_scaled)

anomalies = data_sample[predictions == -1]

print("Detected Anomalies:")
print(len(anomalies))
print(anomalies)
```
